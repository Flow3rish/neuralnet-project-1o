```python
import numpy as np
import pandas as pd
from neural_net.neuralnet import NeuralNet
import neural_net.activation_functions as af
from matplotlib import pyplot as plt
import itertools
import json
from pylab import rcParams
rcParams['figure.figsize'] = 16, 6 # resize figures
plt.style.use('bmh')
```

```python
x = [np.array([np.random.uniform(-1, 10)]) for i in range(1000)]
y = [np.sin(n) for n in x]

training_data = [(x, y) for x, y in zip(x, y)]
plt.scatter(x, y)
```

```python
_x = np.linspace(-1, 10, 1000)
_y = np.sin(_x)
plt.plot(_x, _y)
```

```python
nn = NeuralNet([1, 3, 1], a_functions=[np.tanh, np.tanh], a_functions_prime=[af.tanh_prime, af.tanh_prime])
nn.learn(training_data, 200, 1, .1)
```

```python
outputs = [nn.feed_forward(a)['z'] for a in x]
plt.scatter(x, outputs)
plt.plot(_x, _y, color='r')
```
