```python
import numpy as np
import pandas as pd
from neural_net.neuralnet import NeuralNet
from matplotlib import pyplot as plt
plt.style.use('bmh')
```

```python
x = [np.random.randn(1)*10 - 1 for i in range(4000)]
y = [np.sin(n) for n in x]
x_sig = [NeuralNet.sigmoid(n) for n in x]
y_sig = [NeuralNet.sigmoid(n) for n in y]


training_data = [(x, y) for x, y in zip(x_sig, y_sig)]
```

```python
_x = np.linspace(-10, 10, 1000)
_y = np.sin(_x)
plt.plot(_x, _y)
```

```python
nn = NeuralNet([1, 3, 1])
nn.learn(training_data, 100, 100, .2)
```

```python
outputs = [nn.feed_forward(a) for a in x]
outputs = [NeuralNet.sigmoid_inv(x) for x in outputs]
plt.scatter(x, outputs)
plt.scatter(_x, _y)
```
